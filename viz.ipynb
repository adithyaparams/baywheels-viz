```python
import folium

# Define the latitude and longitude boundaries of Berkeley
berkeley_latitude = (lat_min + lat_max) / 2
berkeley_longitude = (long_min + long_max) / 2

# Create a map centered on Berkeley
map_berkeley = folium.Map(location=[berkeley_latitude, berkeley_longitude], zoom_start=15, tiles='cartodbpositron')

# Define the coordinates of the boundaries
boundary_coordinates = [
    [lat_min, long_min],  # Southwest corner
    [lat_max, long_min],  # Northwest corner
    [lat_max, long_max],  # Northeast corner
    [lat_min, long_max],  # Southeast corner
    [lat_min, long_min],  # Southwest corner (closing the loop)
]

# Create a polygon overlay of the boundary
folium.Polygon(locations=boundary_coordinates, color='blue', fill=True, fill_color='lightblue').add_to(map_berkeley)

# Display the map
map_berkeley
```
